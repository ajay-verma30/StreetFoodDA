```python
countries_with_high_streetfood = df.groupby('Dish Name')['Country'].value_counts()
countries_with_high_streetfood
```

```python
plot_df = countries_with_high_streetfood.rename('Count').reset_index()
plot_df
```

**The above code converts the results in the DataFrame**

```python
plt.figure(figsize=(12, 6))
sns.barplot(data=plot_df, x='Dish Name', y='Count', hue='Country')
plt.title('Street Food Popularity by Country and Dish')
plt.xticks(rotation=90)
plt.show()
```

**We Created a graph to see which street food is more famous and sold widely in the country**

```python
cooking_method_count = df['Cooking Method'].value_counts()
cooking_method_count
```

```python
cooking_method_count.plot(kind='bar')
plt.title('What cooking method do people prefer')
plt.xlabel('Cooking method')
plt.ylabel('Counts')
plt.show()
```

**Based on the above graph it is clear that people widely prefer Grilled followed by Fried and Baked. Whereas the least preference is given to Pan-Fried, Boiled and Stewed**

```python
CookingMethod_Country = df.groupby('Country')['Cooking Method'].value_counts()
```

```python
CookingMethod_Country
```

```python
CookingMethod_Country.plot(kind='bar')
plt.title('What cooking method do people prefer by Country')
plt.xlabel('Cooking method')
plt.ylabel('Counts')
plt.show()
```

```python
df.groupby('Cooking Method')['Typical Price (USD)'].sum().plot(kind='pie',autopct='%1.1f%%',figsize=(10, 10))
plt.title('Cooking Method vs Price')
plt.xlabel('Cooking Method')
plt.ylabel('Price')
plt.show()
```

**We have a clear representation of which cooking method contributes to the more value overall**

```python
df['Vegetarian'].value_counts().plot( kind='pie',
    autopct='%1.1f%%',
    startangle=90,
    figsize=(6, 6),
    colors=['#4caf50', '#f44336'], 
    labels=['Vegetarian', 'Non-Vegetarian'],
    explode=(0.05, 0))
```

**Based on the vegvetarian preferences People on the wide range prefer Vegetarian Food over Non-Vegetarian Food**

```python
grouped = df.groupby(['Cooking Method', 'Vegetarian']).size().unstack(fill_value=0)

grouped.plot(kind='bar', stacked=True, figsize=(10, 6), color=['#f44336', '#4caf50'])
plt.title('Cooking Method vs Vegetarian Preference')
plt.xlabel('Cooking Method')
plt.ylabel('Number of Dishes')
plt.xticks(rotation=45)
plt.legend(title='Vegetarian')
plt.tight_layout()
plt.show()
```

**This graph clearly indicates a strong association between certain cooking methods and vegetarian preference within the dataset. "Grilled" is the dominant method overall, heavily skewed towards non-vegetarian dishes. Conversely, "Assembled," "Baked," and "Boiled/Grilled" are primarily associated with vegetarian dishes. "Fried" and "Stir-Fried" show a more balanced mix of both.**

```python
df['Ingredients List'] = df['Ingredients'].str.split(', ')
exploded = df.explode('Ingredients List')
```

```python
exploded
```

```python
ingredient_counts = exploded['Ingredients List'].value_counts()
ingredient_counts
```

```python
plt.figure(figsize=(10, 6))
ingredient_counts.plot(kind='bar', color='tomato')
plt.title('Top 10 Most Common Ingredients')
plt.xlabel('Ingredient')
plt.ylabel('Frequency')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

**Summary: The most used ingredients is Cheese followed by Spices, Pork, ONions, Flour and Sugar. Whereas the least used are garlic sauce, olive oil, skewers, Herbs and Tahini Sauce**

```python
ingredient_veg = exploded.groupby(['Ingredients List', 'Vegetarian']).size().unstack(fill_value=0)
top_ingredients = ingredient_veg.sum(axis=1).sort_values(ascending=False).head(10)
ingredient_veg_top = ingredient_veg.loc[top_ingredients.index]
```

```python
ingredient_veg_top.plot(kind='bar', figsize=(12, 6), color=['#f44336', '#4caf50'])
plt.title('Top Ingredients by Vegetarian Preference')
plt.xlabel('Ingredient')
plt.ylabel('Count')
plt.xticks(rotation=90)
plt.legend(title='Vegetarian')
plt.tight_layout()
plt.show()
```

**Above chart represents the top 10 ingredients compared by the dietary preferences. By observing the bar chart (Ingredients vs Diet preferences) it is obvious that cheese is widely used in vegetarian diet whereas less in 
non-vegetarian food followed by spices. Whereas when we concentrat on Onions it is clear that it is widely used in non-vegetarian diets then vegetarian diets**
